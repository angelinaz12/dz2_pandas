```python
"""Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity."""


df = pd.read_csv('power.csv')
df.head()
```

```python
"выводим уникальный списко стран, которые содержат в названии ia"
df[ df['country'].str.contains('ia', case=False) ]['country'].unique()
```

```python
"""фильтруем по названием 3 стран из задания"""
filt_cnt = df[ (df['country']=='Lithuania') | (df['country']=='Latvia') | (df['country']=='Estonia')]

filt_cnt.head()
```

```python
"""фильтруем по годам и потреблению"""
df1 = filt_cnt[(filt_cnt['year'] >= 2005) & (filt_cnt['year'] <= 2010) & (filt_cnt['quantity'] > 0)]
df1.head()
```

```python
"""фильтруем по категории,isin фильтрует по нескольким значениям"""
df2 = df1[df1.category.isin([4, 12, 21])]
df2.head()
```

```python
"""суммируем потребление"""
df2['quantity'].sum()
```
